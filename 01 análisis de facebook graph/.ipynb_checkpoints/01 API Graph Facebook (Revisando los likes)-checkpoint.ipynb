Siempre tendr√° 'paging', mas no siempre tendra 'next'.

```python
while 'next' in likes_tmp['paging']:
    likes_tmp = requests.get(likes_tmp['paging']['next']).json()

print('No more likes')
```

```python
def Getlikes(id='me', perimetro='likes', fields=['id','name','created_time','link','description',
                                                   'category','fan_count']):
    lst = []
    likes = graph.get_connections(id, perimetro, fields=','.join(fields))    #get connection de los likes
    
    while True:
        try:
            if 'data' in likes.keys():
                lst += likes['data']
                likes = requests.get(likes['paging']['next']).json()
        
        except KeyError:    #Muy importante
            break
    
    return lst            
```

```python
likes = Getlikes()
print(len(likes))
```

Operaciones con datetime

```python
df1['codmes'] = df1['date'].dt.to_period('M')
df1.head(5).loc[:, ['codmes']]
#df1.head(5)['codmes']
```

```python
#               agrupaci√≥n     |  operaci√≥n  |   extra
likesyear = df1.groupby('year')['id'].count().sort_index(ascending=False)    # likesyear (Series)
likesyear.name = 'likes'
likesyear
```
