¿Existe una persona que trebaje negativo número de horas? No, entonces convierto dichas entradas a NaN.

```python
dfCensus.shape    # df original
```

```python
dfCensus.loc[dfCensus['hours.per.week']<0].shape    # df con filtro 
```

```python
dfCensus.loc[dfCensus['hours.per.week']<0, 'hours.per.week'].nunique()
```

```python
dfCensus.loc[dfCensus['hours.per.week']<0, 'hours.per.week'].sum()
```

```python
len(dfCensus.loc[dfCensus['hours.per.week']<0, 'hours.per.week'])
```

```python
# convertir a NaN

dfCensus.loc[dfCensus['hours.per.week']<0, 'hours.per.week'] = np.nan
```

```python
# Otra forma de convertir usando la instrucción 'where'

dfCensus['hours.per.week'] = np.where(dfCensus['hours.per.week']<0, np.nan, dfCensus['hours.per.week'])
```

```python
# Remplazar por promedio

hours_per_mean = dfCensus['hours.per.week'].mean()
dfCensus['hours.per.week'] = np.where(dfCensus['hours.per.week']<hours_per_mean, 
                                      np.nan, dfCensus['hours.per.week'])
```

```python
# ¿Por qué sigue apareciendo hours.per.week con NaN?
# Porque el filtro, busca aquellos que sean -1

dfCensus.info()
```
